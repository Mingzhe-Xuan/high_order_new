```python
from pymatgen.core import Lattice, Structure

def jarvis_atoms_to_pymatgen(atoms: dict) -> Structure:
    lattice_mat = atoms["lattice_mat"]          # 3x3 lattice matrix
    coords = atoms["coords"]                    # fractional or cartesian
    elements = atoms["elements"]                # species list
    is_cartesian = bool(atoms.get("cartesian", False))

    if len(coords) != len(elements):
        raise ValueError("coords and elements length mismatch")

    lattice = Lattice(lattice_mat)
    structure = Structure(
        lattice=lattice,
        species=elements,
        coords=coords,
        coords_are_cartesian=is_cartesian,
    )
    return structure
```

```python
import json

with open("carbon_24.json", "r") as f:
    data_c = json.load(f)

length = len(data_c)
print("length:", length)
from tqdm import tqdm
import pickle as pkl

print("Start processing...")
for i in tqdm(range(length)):
    row = data_c[i]
    structure = jarvis_atoms_to_pymatgen(row["atoms"])
    row["structure"] = structure

print("Processing completed. Saving to file...")
with open("carbon_24_structures.pkl", "wb") as f:
    pkl.dump(data_c, f)

print("Done.")
```
